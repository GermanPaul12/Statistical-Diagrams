```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from matplotlib.ticker import PercentFormatter
```

```python
data = pd.read_csv('data.csv')
data
```

```python
fig, ax = plt.subplots(figsize=(12,6), )

bar = ax.bar(x=data.States, height=data.Frequency)
ax.set_title('Sales', size=16, fontweight='bold')
ax.set_ylabel('Frequency', fontsize=16)
ax.set_xlabel('States', fontsize=16)
ax.bar_label(bar, label_type='center', color='white', fontsize=14)
ax.tick_params(labelsize=14)
#ax.title.set_size(16) 
```

```python
fig, ax = plt.subplots(figsize=(12,6))

bar = ax.barh(y=data.States, width=data.Frequency)
ax.set_title('Sales', size=16, fontweight='bold')
ax.set_ylabel('Frequency', fontsize=16)
ax.set_xlabel('States', fontsize=16)
ax.bar_label(bar, label_type='center', color='white', fontsize=14)
ax.tick_params(labelsize=14)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

bar = ax.pie(data.Frequency, labels=data.States, autopct='%1.1f%%', textprops={'fontsize': 14})
ax.set_title('Ice Cream Sales Share', size=16, fontweight='bold')
```

```python
df = data 
#sort DataFrame by count descending
df = df.sort_values(by='Frequency', ascending=False)

#add column to display cumulative percentage
df['cumperc'] = df['Frequency'].cumsum()/df['Frequency'].sum()*100
df
```

```python
#define aesthetics for plot
color1 = 'steelblue'
color2 = 'red'
line_size = 8

#create basic bar plot
fig, ax = plt.subplots()
bar = ax.bar(df.States, df['Frequency'], color=color1)

#add cumulative percentage line to plot
ax2 = ax.twinx()
ax2.plot(df.States, df['cumperc'], color=color2, marker="o", ms=line_size)
ax2.yaxis.set_major_formatter(PercentFormatter())

#specify axis colors
ax.tick_params(axis='y', colors=color1)
ax2.tick_params(axis='y', colors=color2)
#styling the plot
ax.set_title('Sales', size=16, fontweight='bold')
ax.set_ylabel('Frequency', fontsize=16)
ax.set_xlabel('States', fontsize=16)
ax.bar_label(bar, label_type='center', color='white', fontsize=14)
ax.tick_params(labelsize=14)
ax2.tick_params(labelsize=14)
```
